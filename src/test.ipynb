```python
import torch
import time

# Chọn device GPU nếu có
if torch.cuda.is_available():
    gpu_device = torch.device("cuda")
elif torch.backends.mps.is_available():
    gpu_device = torch.device("mps")
else:
    print("No GPU available.")
    gpu_device = None

cpu_device = torch.device("cpu")

N = 4000
print(f"Matrix size: {N} x {N}")

# ===== CPU TEST =====
x_cpu = torch.rand((N, N), device=cpu_device)
y_cpu = torch.rand((N, N), device=cpu_device)

print('--- Starting CPU run ---')
start_time = time.time()
cpu_result = torch.matmul(x_cpu, y_cpu)
cpu_time = time.time() - start_time
print(f"CPU time: {cpu_time:.4f} seconds")

# ===== GPU TEST =====
if gpu_device is not None:
    x_gpu = torch.rand((N, N), device=gpu_device)
    y_gpu = torch.rand((N, N), device=gpu_device)

    print('--- Starting GPU run ---')
    
    # Warm-up (quan trọng cho CUDA)
    _ = torch.matmul(x_gpu, y_gpu)
    torch.cuda.synchronize()

    start_time = time.time()
    gpu_result = torch.matmul(x_gpu, y_gpu)
    torch.cuda.synchronize()  # Đợi GPU hoàn thành
    gpu_time = time.time() - start_time

    print(f"GPU time: {gpu_time:.4f} seconds")
    print(f"Speedup: {cpu_time / gpu_time:.2f}x faster")
```
